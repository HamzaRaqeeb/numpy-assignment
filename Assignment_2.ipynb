### Indicators

```python
#set up the indicator I want (just build up the dict if you want more than one)

indicator_ids_1 = ['NE.TRD.GNFS.ZS','NE.IMP.GNFS.CD', 'NE.EXP.GNFS.ZS', 'NY.GDS.TOTL.ZS', 
                   'NY.GDS.TOTL.CN', 'NY.GDS.TOTL.CD'] # indicators of Economy

indicator_ids_2 = ['SL.UEM.TOTL.ZS', 'SE.TER.ENRR', 'SE.ADT.1524.LT.ZS', 'SE.ADT.1524.LT.MA.ZS','SE.ADT.1524.LT.FE.ZS', 
                   'SE.XPD.TOTL.GD.ZS'] # indicators of Education
```

### DataFrames

```python
# Dataframe 1 with indicator_ids_1

my_dataframe_1  = wb.data.DataFrame(indicator_ids_1, country_codes, time=range(2010, 2020), 
                                  skipBlanks=True, columns='series') # most recent 10 years 

my_dataframe_1 = my_dataframe_1.rename(
    columns={'NE.TRD.GNFS.ZS':'Trade (% of GDP)','NE.IMP.GNFS.CD':'Imports(% of GDP)', 
                   'NE.EXP.GNFS.ZS':'Exports(% of GDP)', 'NY.GDS.TOTL.ZS':'Gross domestic savings', 
                   'NY.GDS.TOTL.CN':'Current LCU', 'NY.GDS.TOTL.CD':'Current US$'}
          ,inplace=False) #renaming the indicator ids

my_dataframe_1 = my_dataframe_1.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False) # dropping the null values
```

```python
my_dataframe_1.head(6)
```

```python
# Dataframe 1 with indicator_ids_1

my_dataframe_2  = wb.data.DataFrame(indicator_ids_2, country_codes, time=range(2010, 2020), 
                                  skipBlanks=True, columns='series') # most recent 10 years 

my_dataframe_2  = my_dataframe_2.rename(
    columns={'SE.XPD.TOTL.GD.ZS':'Govt Exp on edu','SE.TER.ENRR':'School Enrollment', 
             'SL.UEM.TOTL.ZS':'Total Unemployment', 'SE.ADT.1524.LT.ZS':'Overall Literacy rate', 
             'SE.ADT.1524.LT.MA.ZS':'Male Literacy rate', 'SE.ADT.1524.LT.FE.ZS':'Female Literacy rate'}
          ,inplace=False) #renaming the indicator ids


my_dataframe_2  = my_dataframe_2.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
# Data filter for finding highest literacy rate among all countries year wise
groups = my_dataframe_2.groupby(by=['time'])
print(groups.apply(lambda g: g[g['Overall Literacy rate'] == g['Overall Literacy rate'].max()]))
```

```python
#Agregrade of data based on Gross domestic savings
averages = my_dataframe_1.groupby('economy')['Gross domestic savings'].mean()
print(averages)
```

```python
#Data Filtering of database 1 based on the mean Expotrs percentage
grouped = my_dataframe_1.groupby('economy')
grouped.filter(lambda x: x['Exports(% of GDP)'].mean() > 35.)
```

```python
#FIltering the data of database 1 based on the std deviation <2
grouped = my_dataframe_1.groupby('economy')
grouped.filter(lambda x: x['Trade (% of GDP)'].std() < 2)
```

```python
#Multilevel grouping by mean of exports and Trade percentages
print(my_dataframe_1.groupby(by=['Exports(% of GDP)', 'Trade (% of GDP)']).mean())
```
